```python
import pandas as pd
```

```python
df = pd.read_csv("merge3.csv")
```

```python
df_final = df.groupby(['tconst', 'originalTitle', 'primaryTitle', 'averageRating', 'numVotes', 'titleType', 'isAdult', 'startYear', 'runtimeMinutes']).agg({'primaryName':set,'primaryProfession':set,'knownForTitles':set,'category':set,'genres':set}).reset_index()
```

```python
df_final['titleType'].unique()
```

```python
df_final.head()
```

```python
mask = df_final[df_final['titleType'].isin(['short', 'movie', 'tvShort', 'tvMovie'])]
```

```python
mask
```

```python
mask = mask[mask['startYear']!="\\N"]
```

```python
mask
```

```python
mask = mask[mask["startYear"].notna()]
mask["startYear"] = mask["startYear"].astype(int)
```

```python
mask_date = mask[mask['startYear'] >= 1945]
```

```python
mask_date
```

```python
mask_vote = mask_date[mask_date["numVotes"] >= 5000]
```

```python
mask_vote
```

```python
tmdb_csv = pd.read_csv("tmdb_csv.csv")
```

```python
df_fin = pd.merge(mask_vote, tmdb_csv, left_on="tconst", right_on="imdb_id")
```

```python
df_fin = df_fin.drop(columns="tconst")
```

```python
df_fin = df_fin.drop(columns="Unnamed: 0")
```

```python
df_fin = df_fin.drop(columns="tagline")
```

```python
df_fin = df_fin.set_index("imdb_id")
```

```python
df_fin.isna().sum()
```

```python
import numpy as np
```

```python
df_fin["release_date"].replace("", np.nan, inplace=True)
df_fin["release_date"].replace(" ", np.nan, inplace=True)
df_fin["release_date"] = df_fin["release_date"].fillna(df_fin["startYear"])
```

```python
df_fin["startYear"] = df_fin["startYear"].astype(str)  # ou int selon besoin
df_fin["release_date"] = df_fin["release_date"].fillna(df_fin["startYear"])
```

```python
df_fin["overview"].replace(["", " "], np.nan, inplace=True)

# Ensuite remplacer les NaN
df_fin["overview"] = df_fin["overview"].fillna("Pas de résumé disponible")
```

```python
df_fin
```

```python
df_fin = df_fin[df_fin["isAdult"] != 1]
```

```python
df_fin = df_fin.drop(columns=["isAdult"])
```

```python
base = "https://image.tmdb.org/t/p/w500/"
df_fin["url_complet"]=df_fin["poster_path"].apply(lambda url: base + str(url) )
```

```python
df_fin =  df_fin.drop(columns=["backdrop_path"])
df_fin =  df_fin.drop(columns=["poster_path"])
```

```python
df_fin.to_csv("df_fin.csv")
```
