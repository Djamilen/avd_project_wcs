```python
from transformers import AutoTokenizer, AutoModel
import torch

#Modèle léger,
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)


def get_emb(texte):
    #Tokenisation + passage dans le modèle,
    inputs = tokenizer(texte, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)

    #Moyenne des embeddings des tokens (mean pooling),
    embeddings = outputs.last_hidden_state.mean(dim=1)

    return embeddings.numpy().flatten()
```

```python
df["corpus"] = (df["primaryTitle"]+ " "+ df.index + " "+ df.index+" "+ " "+ df.index+" "+df["overview"])
```

```python
from tqdm import tqdm
# from tqdm.auto import tqdm  # for notebooks

# Create new `pandas` methods which use `tqdm` progress
# (can use tqdm_gui, optional kwargs, etc.)
tqdm.pandas()
df["vecteurs"] = df["corpus"].progress_apply(get_emb)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommander_film(titre_film,similarity_mat,  n=5):
    titre_film_lower = titre_film.lower()
    
    titres_matches = df[df['primaryTitle'].str.lower().str.contains(titre_film_lower, na=False)]

    if titres_matches.empty:
        print("❌ Film non trouvé dans la base.")
        return


    idx = titres_matches.index[0]
    row_pos = df.index.get_loc(idx)  # Position dans le DataFrame

    sim_scores = similarity_mat.loc[titre_film]
    sim_indices = np.argsort(sim_scores)[0:n+1] # Ignore le film lui-même
    print(f"\n🎬 Films similaires à : {df.iloc[row_pos]['primaryTitle']}\n")
    for i in sim_indices:
        print(f"- {df.iloc[i]['primaryTitle']}")
```
